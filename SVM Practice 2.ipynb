```python
from sklearn.svm import SVC
```

```python
#help(SVC)
```

```python
X = df.drop('Virus Present',axis=1)
```

```python
y = df['Virus Present']
```

```python
SVCmodel = SVC(kernel='linear', C=1000)
```

```python
SVCmodel.fit(X,y)
```

```python
from svm_margin_plot import plot_svm_boundary
```

```python
plot_svm_boundary(SVCmodel,X,y)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
svm = SVC()
```

```python
param_grid= {'C':[100, 10, 1.0, 0.1, 0.001],'kernel':['linear','rbf']}
```

```python
df_grid = GridSearchCV(svm,param_grid)
```

```python
df_grid.fit(X,y)
```

```python
df_grid.best_params_
```

```python
SVCmodel = SVC(kernel='linear', C=100)
SVCmodel.fit(X,y)
```
